## Description
- This is a fictional car insurance company called Car Insured
- They calculate the yearly insurance price based on the current market price of the car
- Given the market price: market_price
- Then the yearly insurance price is 9% of the market_price if it is a male above 26 years old.
- As we see in the decision tree, we have 4 possible price calculations.
- Create a program, which takes input and based on that calculates the correct price for all 4 cases

```python
gender = input("Please enter your gender(M/F): ")
gender = gender.upper()
if gender == "M":
  print("Male")
  age = input("Please enter your age: ")
  age = int(age)
  if age <= 26:
    print("Your age is less than / equal to 26")
    percentage = 23
  else:
    print("Your age is greater than 26")
    percentage = 9
elif gender == "F":
  print("Female")
  car_type = input("What is your car type (sports / non-sports)? ")
  car_type = car_type.upper()
  if car_type == "sports":
    print("You own a sports car")
    percentage = 21
  else:
    print("Not a sports car")
    percentage = 10
else:
  print("Wrong format")
  percentage = None

market_price = input("Please enter the market value of your car: ")
market_price = int(market_price)

Insurance_cost = market_price * percentage / 100

print(f"Annual Insurance Cost: {Insurance_cost}")
```
